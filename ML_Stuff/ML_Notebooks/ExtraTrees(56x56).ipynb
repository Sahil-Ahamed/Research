```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('features_56x56.csv')
```

```python
df
```

```python
X = df.drop(columns=['label'])
Y = df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_scaled = pd.DataFrame(sc.fit_transform(X_train), columns=X_train.columns)
X_test_scaled = pd.DataFrame(sc.fit_transform(X_test), columns=X_test.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
X_train_trf = pca.fit_transform(X_train_scaled)
X_test_trf = pca.transform(X_test_scaled)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
#Moving forward with 9 components as it explains over 98% variance of the data
```

```python
pca = PCA(n_components=9)
X_train_trf = pca.fit_transform(X_train_scaled)
X_test_trf = pca.transform(X_test_scaled)
```

```python
X_train_trf.shape
```

```python
X_test_trf.shape
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create the model
rf_clf = ExtraTreesClassifier(
    n_estimators=200,        # number of trees
    max_depth=None,          # depth of each tree (None = expand until all leaves are pure)
    random_state=42          # for reproducibility
)

# Train the model
rf_clf.fit(X_train_trf, y_train)

# Predict on train and test data
y_train_pred = rf_clf.predict(X_train_trf)
y_test_pred = rf_clf.predict(X_test_trf)

# Evaluate
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, y_test_pred))
print("\nClassification Report:\n", classification_report(y_test, y_test_pred))  # Use y_test_pred here
```

```python
from sklearn.model_selection import cross_val_score
model = ExtraTreesClassifier(
    n_estimators=200,             # Number of trees
    max_depth=10,                 # Limit depth to prevent overfitting
    min_samples_split=10,         # Increase number of samples needed to split a node
    min_samples_leaf=4,           # Increase samples required for leaves
    max_features='sqrt',          # Use square root of features for each split
    bootstrap=True,               # Use bootstrap sampling
    random_state=42
)

# Cross-validation (5-fold) to check model performance
cv_scores = cross_val_score(model, X_train_trf, y_train, cv=5, scoring='recall')

# Print the average cross-validation score
print(f"Cross-Validation Accuracy: {np.mean(cv_scores):.4f} ± {np.std(cv_scores):.4f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Train the model
model.fit(X_train_trf, y_train)

# Predictions
y_train_pred = model.predict(X_train_trf)
y_test_pred = model.predict(X_test_trf)

# Evaluate
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))
print("Testing Accuracy:", accuracy_score(y_test, y_test_pred))
print("\nClassification Report (Test Data):\n", classification_report(y_test, y_test_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("ExtraTrees Classifier - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()
```
